# Topics and Paths

```python
href= occ.VizFolder+"oc_topics_and_paths.html"
heading="Topics and Paths at OpenClassrooms<i> ...et voilà</i>!"

#encoding trick for those who like accents in their titles: 
heading=heading.encode('utf-8').decode('latin')

g = ocg.build_topics_and_paths_graph(height='700px', width='99%', bgcolor="#d5d0f5", font_color="#141414",
                                     heading=heading, show_titles = True,
                                     directed=False,notebook=True,layout=False,show_buttons=False)
g.show(href)
```

```python
# path_id = 410 #architecte logiciel #does not exist anymore as of 2022 12
path_id=592  #Business Intelligence Analyst

# path_id=188  #ingenieur_ia
# path_id=218 #createur entreprise
# path_id=287 #digital marketer
# path_id=319 #technicien info
# path_id=164 # data scientist
path= next(ocd.OC_Paths[ocd.OC_Paths["path_id"].isin([path_id])].iterrows())[1]
path_title=path["path_title"].encode('utf-8').decode('latin')

href= occ.VizFolder+"oc_path_"+str(path_id)+"_projects_courses.html"
 
g = ocg.build_path_projects_courses_graph(path_id=path_id,
                                    max_depth=1,i_read_my_courses_once=False, show_only_requirement_relations = False,  show_relations_max_distance=1, create_pseudo_nodes_over_max_distance=False,
                                     height='680px', width='99%', bgcolor="#d5d0f5", font_color="#141414",
                                     heading=("Projects and Courses for the Path "+path_title),
                                     shape="shad_oc",
                                     directed=True,notebook=True,layout=False,show_titles=True,show_buttons=False)
g.show(href)

# src_dbg= 5800536
# tgt_dbg = 5253451
#ideas
# cut dependency triangles 
# distances between courses via projects skeleton 
```

# Courses

```python
# courses_topic_id = 3 #data
# courses_topic_id = 5 #développement
# courses_topic_id = 13 #autres
courses_topic_id = 0     # allows any node of any topic to be shown
path_topic_id= 0 # allows to show several paths belonging to a given topic. when <>0, the paths are decomposed into projects
courses_language="-"  #"-" allows any language for courses

path_id=188  #ingenieur_ia
# path_id = 410 #architecte logiciel
# path_id = 314 #developpeur frontend
# path_id=0 # allows any path do appear
max_depth=2 #when a path is selected, determines the depth of courses exploration
path_language="-"  #"-" allows any language for paths
show_only_requirement_relations=False

palette="Hybrid" # possibilities = Topic, Hybrid, Dependency
atlas_layout = False # when True, the layout is projected over a globe
show_my_way = True # show "my progression"
connex_paths = False # show only connex paths ?

path_title = "OpenClassrooms Courses"
if path_id!=0:
    path= next(ocd.OC_Paths[ocd.OC_Paths["path_id"].isin([path_id])].iterrows())[1]
    path_title+=" for "+path["path_title"].encode('utf-8').decode('latin')
 #

href= occ.VizFolder+"oc_courses.html"     

filter_options={
    "height"                             : 800,
    "topic"                              : courses_topic_id,
    "language"                           : courses_language,
    "path_topic"                         : path_topic_id,
    "path_language"                      : path_language,
    "path"                               : path_id, 
    "max_depth"                          : max_depth,
    "hide_references"                    : show_only_requirement_relations,
    "atlas_layout"                       : atlas_layout,
    "show_my_way"                        : show_my_way, 
    "palette"                            : palette,
    "connex_paths"                       : connex_paths
}
     
g = ocg.build_courses_graph(bgcolor="#d5d0f5", font_color="#141414",heading=path_title,notebook=True, filter_options=filter_options)  
g.show(href)
```
